```python
list(db['lazada'].aggregate([
  {
    "$unwind": "$responseBody.skus"  #// Unwind the skus array to access each SKU
  },
  {
    "$group": {
      "_id": "$responseBody.itemId",  #// Group by itemId
      "prices": { "$push": "$responseBody.skus.salePrice" },  #// Collect all salePrice values into an array
      "titles": { "$first": "$responseBody.title" },  #// Get the title for the item (assuming it is the same for all SKUs)
      "scraped_timestamps": { "$push": "$scraped_timestamp" }  #// Collect all scraped_timestamp values into an array
    }
  },
  {
    "$project": {
      "_id": 1,  #// itemId
      "prices": 1,  #// All salePrices for the item
      "titles": 1,  #// Title of the product
      "scraped_timestamps": 1  #// All scraped timestamps for the item
    }
  }
]))
```

```python
hist = db['lazada'].aggregate([
  {
    "$unwind": "$responseBody.skus"  # Unwind the skus array to access each SKU
  },
  {
    "$project": {
      "_id": 0,  # Exclude _id field, not needed for the chart
      "itemId": "$responseBody.itemId",  # Include itemId for grouping
      "skuId": "$responseBody.skus.skuId",  # Include skuId for analysis
      "salePrice": "$responseBody.skus.salePrice",  # Get salePrice for each SKU
      "scraped_timestamp": "$scraped_timestamp",  # Use scraped_timestamp for the chart's x-axis
      "title": "$responseBody.title"  # Include the title of the product
    }
  },
  {
    "$sort": {
      "scraped_timestamp": 1  # Sort by scraped_timestamp to ensure chronological order
    }
  },
  {
    "$group": {
      "_id": { "itemId": "$itemId" },  # Group by itemId, skuId
      "price_time_series": { 
        "$addToSet": {  # Use addToSet to avoid duplicate price-timestamp combinations
          "scraped_timestamp": "$scraped_timestamp",  # Timestamp
          "salePrice": "$salePrice"  # Corresponding price
        }
      }
    }
  },
  {
    "$project": {
      "_id": 0,  # Exclude _id field
      "itemId": "$_id.itemId",  # itemId from grouped result
      "skuId": "$_id.skuId",  # skuId from grouped result
      "title": "$_id.title",  # title from grouped result
      "price_time_series": 1  # Array of unique price-time pairs for each SKU
    }
  }
])
```

```python
hist = db['lazada'].aggregate([
  {
    "$unwind": "$responseBody.skus"  # Unwind the skus array to access each SKU
  },
  {
    "$project": {
      "_id": 0,  # Exclude _id field
      "itemId": "$responseBody.itemId",  # Include itemId for grouping
      "skuId": "$responseBody.skus.skuId",  # Include skuId for analysis
      "salePrice": "$responseBody.skus.salePrice",  # Get salePrice for each SKU
      "scraped_timestamp": {
        "$toDate": "$scraped_timestamp"  # Convert scraped_timestamp to Date if it's a string
      },  # Convert scraped_timestamp to Date type
      "title": "$responseBody.title"  # Include the title of the product
    }
  },
  {
    "$project": {
      "_id": 0,  # Exclude _id field
      "itemId": 1,  # Keep itemId for grouping
      "skuId": 1,  # Keep skuId for analysis
      "salePrice": 1,  # Keep salePrice for analysis
      "scraped_timestamp": {
        "$dateToString": { 
          "format": "%Y-%m-%d", 
          "date": "$scraped_timestamp" 
        }
      },  # Truncate scraped_timestamp to yyyy-mm-dd
      "title": 1  # Keep title for the product
    }
  },
  {
    "$sort": {
      "scraped_timestamp": 1  # Sort by scraped_timestamp to ensure chronological order
    }
  },
  {
    "$group": {
      "_id": { "itemId": "$itemId", "scraped_timestamp": "$scraped_timestamp" },  # Group by itemId and date
      "title": { "$first": "$title" },  # Take the first title (same for all SKUs in the group)
      "prices": { "$push": { "skuId": "$skuId", "salePrice": "$salePrice" } }  # Collect skuId and salePrice for each SKU on that day
    }
  },
  {
    "$unwind": "$prices"  # Unwind the prices array so that we can work with individual SKUs
  },
  {
    "$group": {
      "_id": { "itemId": "$_id.itemId", "scraped_timestamp": "$_id.scraped_timestamp" },  # Group by itemId and date
      "title": { "$first": "$title" },  # Keep the title of the product
      "skuId": { "$first": "$prices.skuId" },  # Take the skuId from the first price
      "salePrice": { "$first": "$prices.salePrice" }  # Take the price of the first SKU
    }
  },
  {
    "$project": {
      "_id": 0,  # Exclude _id field
      "itemId": "$_id.itemId",  # itemId from grouped result
      "scraped_timestamp": "$_id.scraped_timestamp",  # scraped_timestamp from grouped result
      "title": 1,  # title from grouped result
      "skuId": 1,  # skuId from grouped result
      "salePrice": 1  # salePrice from grouped result
    }
  }
])    

list(hist)
```

```python
s = db['shopee'].aggregate([
  {
    "$unwind": "$responseBody.data.item"  # Unwind the item array to access each item
  },
  {
    "$project": {
      "_id": 0,  # Exclude _id field
      "itemId": "$responseBody.data.item.itemId",  # Get itemId for grouping
      "title": "$responseBody.data.item.title",  # Get title for each item
      "salePrice": "$responseBody.data.item.price",  # Get price for each item
      "scraped_timestamp": {
        "$toDate": "$scraped_timestamp"  # Convert scraped_timestamp to Date if it's a string
      },  # Convert scraped_timestamp to Date type
    }
  },
  {
    "$project": {
      "_id": 0,  # Exclude _id field
      "itemId": 1,  # Keep itemId for grouping
      "title": 1,  # Keep title for the product
      "salePrice": 1,  # Keep salePrice for analysis
      "scraped_timestamp": {
        "$dateToString": { 
          "format": "%Y-%m-%d", 
          "date": "$scraped_timestamp" 
        }
      },  # Truncate scraped_timestamp to yyyy-mm-dd
    }
  },
  {
    "$sort": {
      "scraped_timestamp": 1  # Sort by scraped_timestamp to ensure chronological order
    }
  },
  {
    "$group": {
      "_id": { "itemId": "$itemId", "scraped_timestamp": "$scraped_timestamp" },  # Group by itemId and date
      "title": { "$first": "$title" },  # Take the first title (same for all items in the group)
      "prices": { "$push": { "salePrice": "$salePrice" } }  # Collect all salePrice for the item on that day
    }
  },
  {
    "$unwind": "$prices"  # Unwind the prices array so that we can work with individual prices
  },
  {
    "$group": {
      "_id": { "itemId": "$_id.itemId", "scraped_timestamp": "$_id.scraped_timestamp" },  # Group by itemId and date
      "title": { "$first": "$title" },  # Keep the title of the product
      "salePrice": { "$first": "$prices.salePrice" }  # Take the first price of the item on that day
    }
  },
  {
    "$project": {
      "_id": 0,  # Exclude _id field
      "itemId": "$_id.itemId",  # itemId from grouped result
      "scraped_timestamp": "$_id.scraped_timestamp",  # scraped_timestamp from grouped result
      "title": 1,  # title from grouped result
      "salePrice": 1  # salePrice from grouped result
    }
  }
])
```

```python
tk = db['tiki'].aggregate([
  {
    "$unwind": "$stock_item"  # Unwind stock_item to access the qty field
  },
  {
    "$project": {
      "_id": 0,  # Exclude _id field
      "id": "$id",  # Include itemId for grouping
      "name": "$name",  # Include name for the product
      "price": "$price",  # Include price for the product
      "stock_qty": "$stock_item.qty",  # Include stock quantity
      "scraped_timestamp": {
        "$toDate": "$scraped_timestamp"  # Convert scraped_timestamp to Date if it's a string
      },  # Convert scraped_timestamp to Date type
    }
  },
  {
    "$project": {
      "_id": 0,  # Exclude _id field
      "id": 1,  # Keep id for grouping
      "name": 1,  # Keep name for the product
      "price": 1,  # Keep price for analysis
      "stock_qty": 1,  # Keep stock_qty for analysis
      "scraped_timestamp": {
        "$dateToString": { 
          "format": "%Y-%m-%d", 
          "date": "$scraped_timestamp" 
        }
      },  # Truncate scraped_timestamp to yyyy-mm-dd
    }
  },
  {
    "$sort": {
      "scraped_timestamp": 1  # Sort by scraped_timestamp to ensure chronological order
    }
  },
  {
    "$group": {
      "_id": { "id": "$id", "scraped_timestamp": "$scraped_timestamp" },  # Group by id and date
      "name": { "$first": "$name" },  # Take the first name (same for all items in the group)
      "price": { "$first": "$price" },  # Take the first price (same for all items in the group)
      "stock_qty": { "$first": "$stock_qty" },  # Take the first stock_qty (same for all items in the group)
    }
  },
  {
    "$project": {
      "_id": 0,  # Exclude _id field
      "id": "$_id.id",  # id from grouped result
      "scraped_timestamp": "$_id.scraped_timestamp",  # scraped_timestamp from grouped result
      "name": 1,  # name from grouped result
      "price": 1,  # price from grouped result
      "stock_qty": 1  # stock_qty from grouped result
    }
  }
])
```

```python
review_on_lazada = db['lazada'].aggregate([
  # Unwind the reviews array to create separate documents
  {
    "$unwind": "$responseBody.reviews"
  },
  
  # Group by itemId
  {
    "$group": {
      "_id": "$responseBody.itemId",
      "reviews": { 
        "$push": "$responseBody.reviews"
      },
      "totalReviews": { 
        "$sum": 1 
      },
      # If ratingCountByScore is an object, we should sum its values
      "ratingCountByScore": { 
        "$first": "$responseBody.ratingCountByScore"
      }    }
  },
  
  # Optional: Sort by number of reviews (descending)
  {
    "$sort": {
      "totalReviews": -1
    }
  }
])
list(review_on_lazada)
```
