```python
import requests
import json
import re
from datetime import datetime
import logging
import yaml

# Configure logging
logging.basicConfig(
    filename="scraping_log.txt",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

# Load URLs from YAML file
def read_urls(urls_file):
    with open(urls_file, 'r') as f:
        config = yaml.load(f, Loader=yaml.FullLoader)
    return config['urls']

# Parse URL to extract item ID, shop ID, and platform
def parse_url(url):
    if re.match(r'^https://shopee.vn/.*[0-9]+\.[0-9]+$', url):
        flag = "shopee"
        url_split = url.split('.')
        itemid = url_split[-1]
        shopid = url_split[-2]
    elif re.match(r'^https:\/\/tiki.vn\/.*[0-9]+\?platform=web&spid=[0-9]+$', url):
        flag = "tiki"
        url_split = re.findall(r".+\/(\d+)\?platform=web&spid=(\d+)", url)
        itemid, shopid = url_split[0]
    elif re.match(r'^https:\/\/tiki.vn\/.*\d+.html.*spid=[0-9]+$', url):
        flag = "tiki"
        url_split = re.findall(r"(\d+).html.*spid=(\d+)$", url)
        itemid, shopid = url_split[0]
    else:
        logging.error(f"Invalid URL format: {url}")
        return None, None, None
    return itemid, shopid, flag

# Tiki Scraper
def fetch_tiki(itemid, shopid):
    url = f'https://tiki.vn/api/v2/products/{shopid}?platform=web&spid={itemid}'
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
    }

    logging.info(f"Starting Tiki scrape for item_id: {itemid} and shop_id: {shopid} at URL: {url}")

    response = requests.get(url, headers=headers)
    timestamp = datetime.now()

    if response.status_code == 200:
        data = response.json()
        data['scraped_timestamp'] = timestamp.isoformat()
        return data
    else:
        raise Exception(f"Failed to fetch data from Tiki API for item_id {itemid} and shop_id {shopid}")

# Main function to process URLs and scrape
def main():
    urls = read_urls("urls.yaml")
    
    for url in urls:
        itemid, shopid, flag = parse_url(url)
        if itemid and shopid:
            try:
                if flag == "tiki":
                    data = fetch_tiki(itemid, shopid)
                    logging.info(f"Successfully fetched data for Tiki URL: {url}")
                    print(data)  # Print or save the data as needed
                elif flag == "shopee":
                    # Add Shopee scraping function here if available
                    logging.info(f"Shopee URL detected, implement scraping for Shopee if needed.")
            except Exception as e:
                logging.error(f"Failed to scrape {flag} URL {url}: {e}")
        else:
            logging.error(f"Skipping URL due to invalid format: {url}")

# Run the main function
if __name__ == "__main__":
    logging.info("Starting scraping process")
    main()
    logging.info("Scraping process completed")
```

```python
import jupyter_dash
from jupyter_dash import JupyterDash
from dash import html, dcc
import plotly.graph_objs as go
from dash.dependencies import Input, Output
import json

# Load data
with open('data/tiki/26140236_2024-11-20.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# Initialize the JupyterDash app
# app = JupyterDash(__name__)
# Initialize the JupyterDash app with specific configuration
app = JupyterDash(__name__, 
                  external_stylesheets=['https://cdnjs.cloudflare.com/ajax/libs/bootstrap/5.3.0/css/bootstrap.min.css'])

# Define the layout
app.layout = html.Div([
    # Header
    html.Div([
        html.H1(data['name'], style={'textAlign': 'center'}),
        html.H3(f"Brand: {data['brand']['name']}", style={'textAlign': 'center'}),
    ], style={'padding': '20px'}),
    
    # Main content
    html.Div([
        # Left column - Product Info
        html.Div([
            html.Div([
                html.H4('Product Details'),
                html.P(f"Price: {data['price']:,} VND"),
                html.P(f"SKU: {data['sku']}"),
                html.P(f"Seller: {data['current_seller']['name']}"),
                
                html.H4('Warranty Information'),
                html.P(f"Warranty Period: {data['specifications'][1]['attributes'][2]['value']}"),
                html.P(f"Warranty Type: {data['specifications'][1]['attributes'][1]['value']}"),
                
                html.H4('Stock Information'),
                html.P(f"Maximum Purchase: {data['stock_item']['max_sale_qty']} units"),
                html.P(f"Minimum Purchase: {data['stock_item']['min_sale_qty']} units"),
            ], style={'backgroundColor': '#f9f9f9', 'padding': '20px', 'borderRadius': '10px'})
        ], style={'width': '30%', 'display': 'inline-block', 'verticalAlign': 'top'}),
        
        # Right column - Visualizations
        html.Div([
            # Price comparison chart
            dcc.Graph(
                figure={
                    'data': [
                        go.Bar(
                            x=['Original Price', 'List Price', 'Current Price'],
                            y=[data['original_price'], data['list_price'], data['price']],
                            marker_color=['rgb(55, 83, 109)', 'rgb(26, 118, 255)', 'rgb(26, 188, 156)']
                        )
                    ],
                    'layout': go.Layout(
                        title='Price Comparison',
                        yaxis={'title': 'Price (VND)'},
                        showlegend=False
                    )
                }
            ),
            
            # Product images carousel
            html.Div([
                html.H4('Product Images'),
                dcc.Slider(
                    id='image-slider',
                    min=0,
                    max=len(data['images'])-1,
                    value=0,
                    marks={i: f'Image {i+1}' for i in range(len(data['images']))},
                    step=1
                ),
                html.Img(id='product-image', style={'maxWidth': '100%', 'marginTop': '20px'})
            ], style={'marginTop': '20px'})
        ], style={'width': '65%', 'display': 'inline-block', 'verticalAlign': 'top', 'marginLeft': '5%'})
    ]),
    
    # Footer with specifications
    html.Div([
        html.H4('Technical Specifications'),
        html.Div([
            html.Ul([
                html.Li('CPU: Intel Core i3-8100 (3.60 GHz)'),
                html.Li('RAM: DDR4 Kingston 8GB/2400'),
                html.Li('GPU: GTX 1050Ti 4GB'),
                html.Li('Storage: SSD 120GB'),
                html.Li('PSU: Xigmatek X450')
            ])
        ], style={'backgroundColor': '#f9f9f9', 'padding': '20px', 'borderRadius': '10px'})
    ], style={'marginTop': '30px'})
])

# Callback to update product image
@app.callback(
    Output('product-image', 'src'),
    [Input('image-slider', 'value')]
)
def update_image(selected_image):
    return data['images'][selected_image]['large_url']

# Run the app
if __name__ == '__main__':
    app.run_server(mode='external', port=8050)
```
